1. What is the difference between a function and a method in Python?
 - A function is a standalone block of code, while a method is a function bound to an object.



```python
# Function example
def greet(name):
    return f"Hello, {name}!"

# Method example (for a string object)
name = "JAIDALI"
print(name.upper())  # 'upper' is a method of the string object
```

2. Explain the concept of function arguments and parameters in Python.
 - In Python, parameters are variables defined in a function's signature, and arguments are the values passed to the function when called.




```python
def greet(name, age):  # 'name' and 'age' are parameters
    return f"Hello {name}, you are {age} years old!"

print(greet("Jaidali", 25))  # "Jaidali" and 25 are arguments
```

3. What are the different ways to define and call a function in Python?
 - In Python, a function can be defined using def or as a lambda function, and can be called by using the function name with arguments.



```python
# Defining a function using def
def add(x, y):
    return x + y

# Calling the function
print(add(2, 3))

# Defining a function using lambda
multiply = lambda x, y: x * y

# Calling the lambda function
print(multiply(2, 3))
```

5. What are iterators in Python and how do they differ from iterables?
 - Iterators are objects that implement the __iter__() and __next__() methods, while iterables are objects that can return an iterator using __iter__().



```python
# Iterable example
numbers = [1, 2, 3]  # List is an iterable

# Iterator example
iterator = iter(numbers)  # Create an iterator from the list
print(next(iterator))  # Outputs 1
print(next(iterator))  # Outputs 2
print(next(iterator))  # Outputs 3
```

6. Explain the concept of generators in Python and how they are defined.
 - Generators in Python are functions that use the yield keyword to produce a series of values lazily, instead of returning them all at once.



```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator
gen = count_up_to(3)
print(next(gen))  # Outputs 1
print(next(gen))  # Outputs 2
print(next(gen))  # Outputs 3
```

7. What are the advantages of using generators over regular functions?
 - Generators are more memory-efficient than regular functions because they yield values one at a time instead of returning all values at once.



```python
# Regular function (returns a full list)
def create_numbers(n):
    return [i for i in range(1, n+1)]

# Generator (yields numbers one at a time)
def generate_numbers(n):
    for i in range(1, n+1):
        yield i

# Using the regular function
numbers_list = create_numbers(3)
print(numbers_list)  # Outputs [1, 2, 3]

# Using the generator
gen = generate_numbers(3)
print(next(gen))  # Outputs 1
print(next(gen))  # Outputs 2
print(next(gen))  # Outputs 3
```

8. What is a lambda function in Python and when is it typically used?
 - A lambda function is a small anonymous function defined with the lambda keyword, typically used for short-term operations or as arguments to higher-order functions.



```python
# Lambda function to add two numbers
add = lambda x, y: x + y

# Using the lambda function
print(add(2, 3))  # Outputs 5
```

9. Explain the purpose and usage of the map() function in Python.
 - The map() function in Python applies a given function to all items in an iterable and returns an iterator with the results.



```python
# Function to square a number
def square(x):
    return x ** 2

# Using map to apply 'square' to a list of numbers
numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)

# Convert the result to a list and print
print(list(squared_numbers))  # Outputs [1, 4, 9, 16]
```

10. What is the difference between map(), reduce(), and filter() functions in Python?
 - map() applies a function to each item in an iterable, reduce() applies a function cumulatively to reduce an iterable to a single value, and filter() filters items based on a condition.



```python
from functools import reduce

# map example (squaring each number)
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))  # Outputs [1, 4, 9, 16]

# reduce example (calculating the product of all numbers)
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Outputs 24

# filter example (filtering even numbers)
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Outputs [2, 4]
```

11. Using pen & Paper write the internal mechanism for sum operation using  reduce function on this given
list:[47,11,42,13]
 - The reduce() function applies the sum operation cumulatively on the list [47, 11, 42, 13], starting from the first two elements, then applying it to the result with the next element, and so on.



```python
from functools import reduce

numbers = [47, 11, 42, 13]
sum_result = reduce(lambda x, y: x + y, numbers)
print(sum_result)  # Outputs 113
```
